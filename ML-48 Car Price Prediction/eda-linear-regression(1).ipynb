```python
''' dropping carID column '''
df.drop(['car_ID'], axis=1, inplace=True)
```

```python
''' value counts of fueltype '''
ft_label = df.fueltype.value_counts()

''' barplot '''
plt.figure(figsize=(8, 5))
sns.barplot(ft_label.index, ft_label);
plt.xlabel('Fuel Type', fontsize=15)
plt.ylabel('Count', fontsize=15);
```

```python
''' value_counts of carbody '''
cb_label = df.carbody.value_counts()

''' barplot '''
plt.figure(figsize=(8, 5))
sns.barplot(cb_label.index, cb_label);
plt.xlabel('Car Body', fontsize=15)
plt.ylabel('Count', fontsize=15);
```

```python
df.head()
```

```python
''' extracing car brand '''
df['Car_Brand'] = df.CarName.apply(lambda x: x.split()[0])
```

```python
''' value_counts of car_brand '''
cb = df.Car_Brand.value_counts()[:5]

''' barplot '''
plt.figure(figsize=(8, 5))
sns.barplot( cb);
plt.xlabel('Car Brand', fontsize=15)
plt.ylabel('Count', fontsize=15);
```

```python
brand_map = {
    'toyouta':'toyota',
    'Nissan':'nissan',
    'maxda':'mazda',
    'vokswagen':'volkswagen',
    'vw':'volkswagen',
    'porcshce':'porsche'
}
```

```python
df['Car_Brand'] = df.Car_Brand.apply(lambda s: brand_map[s] if s in brand_map else s)
```

```python
df['Car_Brand'].value_counts()
```

```python
''' dropping CarName column '''
df.drop('CarName', axis=1, inplace=True)
```

```python
df.head()
```

```python
''' histogram plot pf price '''
plt.figure(figsize=(10, 5))
sns.histplot(df['price'])
plt.title('Histogram plot of price', fontsize=20)
plt.xlabel('price', fontsize=15)
plt.ylabel('count', fontsize=15);
```

```python
''' boxplot of price'''
plt.figure(figsize=(10, 5))
sns.boxplot(df['price'])
plt.title('Box plot of  price', fontsize=20)
plt.xlabel('price', fontsize=15);
```

```python
''' count plot of symboling '''
plt.figure(figsize=(10, 5))
sns.countplot(df['symboling']);
plt.xlabel('symboling', fontsize=15)
plt.ylabel('count', fontsize=15);
```

```python
''' boxplot '''
plt.figure(figsize=(10, 5))
sns.boxplot(y='price', x='symboling', data=df)
plt.xlabel('symboling', fontsize=15)
plt.ylabel('price', fontsize=15);
```

```python
''' boxplot '''
plt.figure(figsize=(10, 5))
sns.boxplot(y='price', x='fueltype', data=df)
plt.xlabel('fueltype', fontsize=15)
plt.ylabel('price', fontsize=15);
```

```python
''' boxplot '''
plt.figure(figsize=(10, 5))
sns.boxplot(y='price', x='aspiration', data=df)
plt.xlabel('aspiration', fontsize=15)
plt.ylabel('price', fontsize=15);
```

```python
''' correlation matrix '''
plt.figure(figsize=(20, 15))
sns.heatmap(df.corr(), annot=True)
plt.title('Confusion Matrix', fontsize=20);
```

```python
c_cols = [feature for feature in df.columns if df[feature].dtype == 'int64' or 'object']
```

```python
c_cols
```

```python
n_cols = 4
n_rows = int(np.ceil(len(c_cols) / n_cols))
t_cols = len(c_cols)

''' boxplot '''
fig, axs = plt.subplots(ncols=n_cols, nrows=n_rows, sharey = True, figsize=(20,20))
for i, col in enumerate(c_cols):
    row_idx = i // n_cols
    col_idx = i % n_cols
    sns.boxplot(y='price', x=col, data=df, ax=axs[row_idx, col_idx])
plt.tight_layout()
plt.show()
```
